Extracting the Audio from Video

```python
pip install moviepy
```

```python
import moviepy.editor
```

```python
from moviepy.video.io.ffmpeg_tools import ffmpeg_extract_audio
from moviepy.editor import VideoFileClip

video_file = "one.mp4"


audio_file = "output_audio.mp3"

start_time = 15
end_time = 80


video_clip = VideoFileClip(video_file)


subclip = video_clip.subclip(start_time, end_time)


subclip.audio.write_audiofile(audio_file, codec='mp3')
```

Now that we have the audio file of the speaker, we are now using this audio file to get the transcription of the file.

```python
pip install openai
```

```python
pip install openai-whisper
```

```python
import whisper
```

```python
model = whisper.load_model("base")
result = model.transcribe("output_audio.mp3")
with open("transcription_one.txt","w") as f:
  f.write(result["text"])
```
